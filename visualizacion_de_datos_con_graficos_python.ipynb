## Matplotlib

```python
import pandas as pd
```

```python
df_inmigrantes = pd.read_csv('/content/inmigrantes_canada.csv')
df_inmigrantes.head()
```

```python
# mostrar la cantidad de filas y columnas que tiene mi df 195 filas y 38 columnas
df_inmigrantes.shape
```

```python
df_inmigrantes.info()
```

```python
# Para modificar el indice, que no aparezca el n° sino el pais como indice
df_inmigrantes.set_index('Pais', inplace=True)
df_inmigrantes.head()
```

```python
# generar una variable con una lista con los strig con las columnas que son los años
anos=list(map(str,range(1980, 2014))) # incluye el año 1980 y excluye el 2014 osea llega hasta el 2013
anos
```

```python
# para poder hacer el filtro especifico
# me muestra una series con los años y la cantidad de inmigrantes colombianos en canada por año
colombia = df_inmigrantes.loc['Colombia', anos]
colombia
```

```python
# creo un DF con un diccionario donde me mostrara en una lista los años
# llamo a la series creada anteriormente llamada colombia y le digo que el indica es el año y los valores son las cantidades de inmigrantes

# colombia.index.tolist(): Esto convierte el índice de tu DataFrame o Serie colombia a una lista de strings.
# colombia.values.tolist(): Esto convierte los valores (probablemente tipo np.int64) en una lista.

col_dict = {'Año': colombia.index.tolist(),
           'Inmigrantes': [int(x) for x in colombia.values.tolist()]} # tengo que hacer esta transformación para que no aparezca la lista con np.int64
col_dict
```

```python
# Para crear un Data Frame con los datos del diccionario anterior
datos_colombia = pd.DataFrame(col_dict)
datos_colombia
```

```python
# biblioteca para hacer gráficos
import matplotlib.pyplot as plt
```

```python
plt.plot(datos_colombia['Año'], datos_colombia['Inmigrantes'])
plt.show()
```

```python
# xticks este comando es para poner los años como str en y y los genero en rangos para ser más legible el gráfico
plt.plot(datos_colombia['Año'], datos_colombia['Inmigrantes'])
plt.xlabel('Año')
plt.ylabel('Inmigrantes')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2013'])
plt.show()
```

```python
plt.figure(figsize=(6,4)) # tamaño de la figura en pulgadas: 6 de ancho * 4 de alto
plt.plot(datos_colombia['Año'], datos_colombia['Inmigrantes']) # para generar un gráfico de líneas plot
plt.xlabel('Año') # etiqueta eje x
plt.ylabel('Inmigrantes') # etiqueta eje y
plt.title('Inmigrantes Colombianos en Canadá')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2013'])
plt.show() # este el el último comando, después de esto no se lle nada más
```

**Analizar más paises: Colombia, Brasil, Argentina, Chile

```python
fig, axs = plt.subplots(2,2, figsize=(10,6)) # 2 espacios * 2 gáficos
fig.subplots_adjust(hspace=0.5, wspace=0.5) # espacio entre gráficos h de horizontal y w de ancho
fig.suptitle('Inmigrantes a Canadá por País')

# selecciono a traves del .loc los paises del DF original y llamo a la lista anos que generer más arriba
axs[0,0].plot(df_inmigrantes.loc['Colombia',anos])
axs[0,0].set_title('Colombia')
axs[0,1].plot(df_inmigrantes.loc['Brasil',anos])
axs[0,1].set_title('Brasil')
axs[1,0].plot(df_inmigrantes.loc['Chile',anos])
axs[1,0].set_title('Chile')
axs[1,1].plot(df_inmigrantes.loc['Argentina',anos])
axs[1,1].set_title('Argentina')


# propiedad .flat esto es para iterar en cada una de las gráficas
for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # MultipleLocator sirve para definir la separación entre marcas (ticks) en un eje de un
                                                       # gráfico (x o y). Es decir, le dices: “Quiero que cada tick esté cada n unidades”.
    ax.set_xlabel('Año')
    ax.set_ylabel('N° Inmigrantes')

plt.tight_layout()
#pertenece a Matplotlib y se usa para ajustar automáticamente los elementos de una figura (gráfica)
# para que no se sobrepongan ni queden cortados.
plt.show()
```

```python
# para generar un valor mínimo y máximo para cada gráfico y no se preste a confusión
ymin=0
ymax=7000

fig, axs = plt.subplots(2,2, figsize=(10,6)) # 2 espacios * 2 gáficos
fig.subplots_adjust(hspace=0.5, wspace=0.5) # espacio entre gráficos h de horizontal y w de ancho
fig.suptitle('Inmigrantes a Canadá por País')

# selecciono a traves del .loc los paises del DF original y llamo a la lista anos que generer más arriba
axs[0,0].plot(df_inmigrantes.loc['Colombia',anos])
axs[0,0].set_title('Colombia')
axs[0,1].plot(df_inmigrantes.loc['Brasil',anos])
axs[0,1].set_title('Brasil')
axs[1,0].plot(df_inmigrantes.loc['Chile',anos])
axs[1,0].set_title('Chile')
axs[1,1].plot(df_inmigrantes.loc['Argentina',anos])
axs[1,1].set_title('Argentina')
# entrar a un array multidimensional y traer un array unidimensional
for ax in axs.ravel():
    ax.set_ylim(ymin,ymax)
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.set_xlabel('Año')
    ax.set_ylabel('N° Inmigrantes')
    ax.grid()

plt.tight_layout()
plt.show()
```

Pero además, según la documentación de la biblioteca Matplotlib, podemos aplicar diferentes estilos para permitirnos adaptar las visualizaciones de acuerdo con nuestras necesidades. Para saber qué estilos están disponibles, podemos imprimir plt.style.available, donde style es un submódulo de la biblioteca Matplotlib utilizado para definir estilos de visualización:

```python
print(plt.style.available)
```

¡Podemos explorar uno de estos estilos para mejorar aún más nuestras visualizaciones!

Antes de aplicar un estilo, es importante saber que cada vez que la biblioteca Matplotlib es importada en un notebook, define una configuración de tiempo de ejecución que incluye los estilos predeterminados para cada elemento de trazado creado.

Por lo tanto, para evitar que el estilo se aplique a todos los gráficos trazados en el mismo notebook, podemos utilizar un código que crea una copia de las configuraciones predeterminadas de trazado de la biblioteca Matplotlib y las asigna a la variable IPython_default. Esto puede ser útil para almacenar y reutilizar las configuraciones predeterminadas de trazado o para restaurarlas después de haber sido modificadas:

IPython_default = plt.rcParams.copy()


Uno de los estilos disponibles se basa en un sitio de noticias y análisis de datos llamado FiveThirtyEight, que cubre temas como política, economía, cultura, ciencia y deportes.

Para utilizar el estilo 'fivethirtyeight', usamos el siguiente código:

plt.style.use('fivethirtyeight')

A continuación, podemos crear la figura, que tendrá este nuevo estilo incorporado:

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(datos_col['Año'], datos_col['Inmigrantes'])

ax.set_title('Inmigración de colombianos hacia Canadá\n1980 a 2013', fontsize=20, loc='left')

ax.set_ylabel('Número de Inmigrantes', fontsize=14)

ax.set_xlabel('Año', fontsize=14)

ax.yaxis.set_tick_params(labelsize=12)

ax.xaxis.set_tick_params(labelsize=12)

ax.xaxis.set_major_locator(plt.MultipleLocator(5))

plt.show()

Los gráficos creados con el estilo FiveThirtyEight son atractivos y poseen una estética limpia y minimalista, con líneas más gruesas y colores vibrantes.

Bueno, ¿y si queremos trazar gráficos sin este estilo después de haberlo definido en el notebook? Podemos redefinir los parámetros utilizando plt.rcParams.update y pasándole la variable IPython_default que creamos anteriormente con las configuraciones predeterminadas:

plt.rcParams.update(IPython_default);

Además, podemos aplicar un estilo solo a un bloque de código específico, ya que el paquete de estilo proporciona un administrador de contexto para limitar sus cambios a un ámbito específico. Para aislar tus cambios de estilo, puedes escribir el código dentro de un contexto with de la siguiente manera:

with plt.style.context('fivethirtyeight'):

  fig, ax = plt.subplots(figsize=(8, 4))

  ax.plot(datos_col['Año'], datos_col['Inmigrantes'])

  ax.set_title('Inmigración de colombianos hacia Canadá\n1980 a 2013', fontsize=20, loc='left')

  ax.set_ylabel('Número de Inmigrantes', fontsize=14)

  ax.set_xlabel('Año', fontsize=14)

  ax.yaxis.set_tick_params(labelsize=12)

  ax.xaxis.set_tick_params(labelsize=12)

  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  
  plt.show()

```python
sudamerica = df_inmigrantes.query('Region == "América del Sur"')
sudamerica.head(10)
```

```python
colores=['royalblue','orange','forestgreen','orchid','olive','purple','gray','navy','teal','yellow','red','blue']
fig, ax = plt.subplots(figsize=(10,3))
ax.bar(sudamerica.index, sudamerica['Total'], color=colores)
ax.set_title('Inmigración de Sudaméricanos hacia Canada durante el périodo\n1980 - 2013', loc='left', fontsize=16)
ax.set_xlabel('')
ax.set_ylabel('N° de Inmigrantes', fontsize=12)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
# este gráfico de sudamericanos está desordenado
colores=['royalblue','orange','forestgreen','orchid','olive','purple','gray','navy','teal','yellow','red','blue']

fig, ax = plt.subplots(figsize=(10,3))
ax.barh(sudamerica.index, sudamerica['Total'], color=colores)
ax.set_title('Inmigración de Sudaméricanos hacia Canada durante el périodo\n1980 - 2013', loc='left', fontsize=16)
ax.set_xlabel('N° de Inmigrantes', fontsize=12)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
#creo un nuevo df con sort para ordenar de forma ascendente (es un método de lista que ordena la lista en el lugar)
sudamerica_sorted = sudamerica.sort_values('Total', ascending=True)

colores=['royalblue','orange','forestgreen','orchid','olive','purple','gray','navy','teal','yellow','red','blue']

fig, ax = plt.subplots(figsize=(10,3))
ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Inmigración de Sudaméricanos hacia Canada durante el périodo\n1980 - 2013', loc='left', fontsize=16)
ax.set_xlabel('N° de Inmigrantes', fontsize=12)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
colores = []
for pais in sudamerica_sorted.index:
  if pais == 'Colombia':
    colores.append('green')
  else:
    colores.append('gray')

fig, ax = plt.subplots(figsize=(10,3))

# Estos 2 comandos son para que se hagan invisibles las líneas del fondo y los ejes
ax.set_frame_on(False)
ax.xaxis.set_visible(False)

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Inmigración de Columbianos hacia Canadá durante el périodo\n1980 - 2013', loc='left', fontsize=16)
#ax.set_xlabel('N° de Inmigrantes', fontsize=12)
ax.set_ylabel('')
#ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=10)

for i,j in enumerate(sudamerica_sorted['Total']):
  ax.text(j+20,i, str(j), va='center', ha='left', color='black', fontsize=9) # esto es para que aparezcan los valores al lado de las barras

fig.savefig('/content/inmigrantes_sudamericanos_canada.png', transparent=False,
            dpi = 300, bbox_inches='tight')
plt.show()
```

```python
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  # Estos 2 comandos son para que se hagan invisibles las líneas del fondo y los ejes
  ax.set(xticklabels=[]) # esto sirve para no aparezca el n° de inmigrantes

  ax = sns.barplot(data=top_10, x='Total', y= top_10.index,  hue=top_10.index, orient='h',
                   palette=palette , legend=False) #  Con hue=top_10.index le dices a Seaborn:“Colorea usando esta categoría”.
                                                                    # Y legend=False evita que te dibuje una leyenda repetitiva.
  ax.set_title('Top 10 países con mayor migración hacia Canadá\n1980-2013',
        loc = 'left', fontsize = 16) # loc = 'left' significa que el tirulo estara ubicado hacia la izquierda, fontsize = 16 (tamaño letra)
  ax.set_xlabel('N° de Inmigrantes', fontsize=14)
  ax.set_ylabel('')

  sns.despine() # quita el marco derecho y el de arriba

  for i,j in enumerate(top_10['Total']):
    ax.text(j+20,i, str(j), va='center', ha='left', color='black', fontsize=9) # esto es para que aparezcan los valores al lado de las barras

plt.show()
```
